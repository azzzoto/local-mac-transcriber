```python
import os
import glob
from os.path import join
from tqdm import tqdm

list_of_files = glob.glob("*.m4a")
directory_names = [item[:-4] for item in list_of_files]
```

```python
import subprocess

def split_files(list_of_files=list_of_files, chunk_length=1800):
    for filename in tqdm(list_of_files):
        source_path = join("/content", filename)
        directory_name = filename[:-4]
        os.mkdir(directory_name)
        command = [
            "ffmpeg",
            "-i", source_path,
            "-f", "segment",
            "-segment_time", str(chunk_length),
            "-c", "copy",
            f"{directory_name}/{directory_name}_part%03d.aac"
        ]

        try:
            result = subprocess.run(command, capture_output=True, text=True, check=True)
            print(result.stdout)
        except subprocess.CalledProcessError as error:
            print(f"Error: {error.stderr}")
```

```python
split_files()
```
